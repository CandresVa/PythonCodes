```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
#Only numbers
def validation(value):
    while True:
        try:
            return float(value)
        except ValueError:
            value=input("ATENTION: Only numbers input: ")
```

```python
print('Circular duct:')
R=(input('Radius :'));R=validation(value=R)


print('Duct divitions:')
Refi=(input('Divitions d (dxd) :')); Refi=validation(value=Refi)


#Input parameters:
#Es sobre un cuadrado de (2R)x(2R)
Delta_x = 2*R/Refi #m
Delta_y = 2*R/Refi #m  
print(Delta_x)

#Calculos complementarios
No_nodes_x =int(2*R/Delta_x +1)
No_nodes_y =int(2*R/Delta_y +1)

# Solución inicial: malla -> w(k, i,j )
Max_iter=100
Nodes_x=2*No_nodes_x
w= np.zeros((Max_iter,2*No_nodes_x,2*No_nodes_y))

dp=-1000
```

```python
for k in range(1, Max_iter-1, 1):
    for i in range(1, 2*No_nodes_x, 1):
        for j in range(1, 2*No_nodes_y, 1):
            comp=((R-i)**2+(R-j)**2)**0.5
            if comp<R:
                #w[j,i] = 10
                w[k,j,i] = (-dp*Delta_x**2 * Delta_y**2 + Delta_y**2 *(w[k,j,i-1] + w[k-1,j,i+1]) + Delta_x**2 * (w[k,j-1,i] + w[k-1,j+1,i]) )  /  (2*(Delta_x**2 + Delta_y**2))

        #for i in range(1, No_nodes_y-1, 1): # w[k,j,i]
            #w[k,j,i] = (dp*Delta_x**2 * Delta_y**2 + Delta_y**2 *(w[k,j,i-1] + w[k-1,j,i+1]) + Delta_x**2 * (w[k,j-1,i] + w[k-1,j+1,i]) )  /  (2*(Delta_x**2 + Delta_y**2))
            #w[k,j,i] = 0.5*((Delta_x**2*Delta_y**2)/(Delta_x**2+Delta_y**2))*(((w[k,j,i+1]+w[k,j,i-1])/Delta_x**2)+((w[k,j,i+1]+w[k,j,i-1])/Delta_y**2)-dp)
            #w[k,j,i] = (-dp*Delta_x**2 * Delta_y**2 + Delta_y**2 *(w[k,j,i-1] + w[k-1,j,i+1]) + Delta_x**2 * (w[k,j-1,i] + w[k-1,j+1,i]) )  /  (2*(Delta_x**2 + Delta_y**2))


        #CODIGO PARAGRAFICAR

plt.title(f"Perfil de velocilad")
plt.xlabel("x")
plt.ylabel("y")
plt.pcolormesh(w[k], cmap=plt.cm.jet, vmin=0, vmax=np.max(w[k]))

plt.colorbar()
```
